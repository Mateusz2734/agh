### 1. Ilu użytkowników lokalnych jest zdefiniowanych w systemie?

```cat /etc/passwd | wc -l```

### 2. Jak w systemie opisany został użytkownik, którego numer identyfikacyjny wynosi 14?

```cat /etc/passwd | awk -F ':' '{if ($3 == 14) print $5}'```

### 3. Ilu jest użytkowników w systemie, których grupa podstawowa jest Twoją grupą?

```cat /etc/passwd | awk -F ':' "{if (\$4 == $(id -g)) print \$0}" | wc -l```

### 4. Ilu użytkowników zdefiniowanych w systemie używa jako podstawowego interpretera poleceń interpretera Bash?

```cat /etc/passwd | awk -F ':' "{print \$7}" | grep -E "^/bin/bash$" | wc -l```

### 5. Ilu użytkowników w systemie posiada numer identyfikacyjny UID większy od 12000?

```awk -F ':' '{if ($3 > 12000) print $0}' /etc/passwd | wc -l```

### 6. Jak wylistować zawartość kilku katalogów na raz (w jednej komendzie)?

```ls katalog1 katalog2 katalog3```

### 7. Jak wypisać za pomocą jednej komendy echo cyfry od 0 do 5, każdą w osobnej linii?

```echo -e \\b{1..5}\\n```
lub
```echo -e "1\n2\n3\n4\n5"```
lub ```seq 1 5```

### Ile plików i katalogów znajduje się w katalogu głównym (/)?

```ls / | wc -l```

### Ilu użytkowników jest aktualnie zalogowanych?

```who |wc -l```

### Ile słów liczy plik /usr/include/utime.h?

```cat /usr/include/utime.h | wc -w```
