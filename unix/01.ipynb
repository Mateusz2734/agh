### 1. Ilu użytkowników lokalnych jest zdefiniowanych w systemie?

```cat /etc/passwd | wc -l```

### 2. Jak w systemie opisany został użytkownik, którego numer identyfikacyjny wynosi 14?

```cat /etc/passwd | awk -F ':' '{if ($3 == 14) print $5}'```
