Proszę napisać skrypt Bash wg poniższej specyfikacji i załączyć go jako odpowiedź na pytanie.

# Funkcjonalność
Skrypt powinien wyświetlić ilość pamięci RSS zajmowanej przez procesy jednego lub wielu wskazanych przy wywołaniu użytkowników.
Informacje o zajętości pamięci powinny być wypisywane dla każdego poprawnie wskazanego użytkownika w jednej osobnej linii. W każdej linii powinna być zawarta:

- nazwa użytkownika,

- łączna ilość pamięci RSS zajmowanej przez wszystkie procesy, których właścicielem jest ten użytkownik.

Poszczególne pola powinny być oddzielone spacjami. Można przyjąć dowolne jednostki, w których będzie wyświetlana informacja o zajętości pamięci. Jeśli użytkownik wskazany na liście nie istnieje lub nie można dla niego sprawdzić liczby procesów, to powinien zostać wypisany komunikat o błędzie, a skrypt powinien zakończyć swoje działanie. Skrypt nie powinien wypisywać na standardowe wyjście żadnych innych informacji, które nie zostały wymienione w niniejszej specyfikacji.

Przykładowe wywołanie skryptu skrypt.sh mające na celu wyświetlenie zajętości pamięci dla użytkowników user i root będzie następujące:

```./skrypt.sh user root```

Przykładowy wynik działania dla powyższego wywołania może być następujący:

```user 40560
root 657532```

# Argumenty wejściowe i opcje wywołania
Podczas wywołania skryptu należy podać listę użytkowników, dla których ma być wyświetlona zajętość pamięci. Jeśli nie zostanie podany żaden argument, skrypt powinien wyświetlić komunikat o błędnym wywołaniu i zakończyć swoje działanie.

# Wartości zwracane
Skrypt przy zakończeniu działania powinien zwracać wartość:

- 0, jeśli jego wywołanie i wykonanie przebiegło bez błędów;
- liczba inna niż zero, jeśli nie wskazano żadnych użytkowników bądź w pozostałych przypadkach, np. przy istotnych błędach wykrytych w trakcie działania skryptu.

Przy wartości zwracanej innej niż 0 należy wypisać krótki komunikat tekstowy o błędzie.

```sh
#!/bin/bash

function calc_sum_and_print {
    command=$(ps --no-headers -eo user,rss | sort | grep "^$1 " | awk '{SUM+=$2}END{print SUM}')

    if [[ ! -n $command ]] ; then
        echo "Error: User $1 does not exist"
        exit 1

    else
        echo "$1 $command"
    fi
}

if (( $# == 0)) ; then
    echo "Error: 0 users specified"
    exit 2
fi

while (($# > 0))
do
    calc_sum_and_print $1
    shift
done
exit 0
```
