```python
def max_spojny_podciag(T):
    n = len(T)
    output = []
    i = j = 0
    while j < n - 1:
        if T[j] < T[j+1]:
            j += 1
        else:
            if j - i >= 2:
                output.append([T[i], T[j]])
            i = j + 1
            j = i
    return output

print(max_spojny_podciag([1,2,3,2,1,3,5,6]))
```
