### Funkcje pomocnicze

```python
from math import isqrt, log

def is_prime(num):
    if num <= 1:
        return False
    if num == 2 or num == 3:
        return True
    if num%2 == 0 or num%3 == 0:
        return False
    i = 5
    while i < isqrt(num) + 1:
        if num % i == 0:
            return False
        i += 2
        if num % i == 0:
            return False
        i += 4
    return True
```

```python
def only_odd(num):
    while num > 0:
        if (num % 10) % 2 == 0:
            return False
        num //= 10
    return True

def only_odd_in_row(T):
    n = len(T)
    for i in range(n):
        flag = False
        for j in range(n):
            if only_odd(T[i][j]):
                flag = True
        if not flag:
            return False
    return True
        
tab = [[22, 33, 44],
       [22, 33, 44],
       [22, 33, 44]]

print(only_odd_in_row(tab))
```

```python
def one_even(num):
    while num > 0:
        if (num % 10) % 2 == 0:
            return True
        num //= 10
    return False

def at_least_one_even(T):
    n = len(T)
    main_flag = False
    for i in range(n):
        secondary_flag = True
        for j in range(n):
            if not one_even(T[i][j]):
                secondary_flag = False
        if secondary_flag:
            main_flag = True
    return main_flag


tab = [[22, 34, 44],
       [22, 33, 44],
       [22, 33, 44]]
print(at_least_one_even(tab))
```

### Zadanie 10. Napisać funkcję która dla tablicy $T[N][N]$, wypełnionej liczbami całkowitymi, zwraca wartość True w przypadku, gdy w każdym wierszu i każdej kolumnie występuje co najmniej jedno $0$ oraz wartość False w przeciwnym przypadku.

```python
def column_row_zero(T):
    n = len(T)
    col = [False for _ in range(n)]
    row = [False for _ in range(n)]
    for i in range(n):
        for j in range(n):
            if T[i][j] == 0:
                row[i] = True
                col[j] = True
    return not ((False in row) or (False in col))

tab = [[1, 0, 1],
       [0, 1, 0],
       [1, 1, 0]]
print(column_row_zero(tab))        
```

### Zadanie 15. Dana jest tablica $T[N][N]$, wypełniona liczbami naturalnymi. Proszę napisać funkcję, która odpowiada na pytanie, czy w tablicy istnieje wiersz, w którym każda liczba zawiera co najmniej jedną cyfrę będącą liczbą pierwszą?

```python


def one_prime(num):
    while num > 0:
        if is_prime(num % 10):
            return True
        num //= 10
    return False

def one_prime_tab(T):
    n = len(T)
    main_flag = False
    for i in range(n):
        secondary_flag = True
        for j in range(n):
            if not one_prime(T[i][j]):
                secondary_flag = False
        if secondary_flag == True:
            main_flag = True
    return main_flag
        
tab = [[4, 4, 4],
       [3, 2, 3],
       [4, 4, 4]]
print(one_prime_tab(tab)) 
```

### Zadanie 16. Dana jest tablica $T[N][N]$, wypełniona liczbami naturalnymi. Proszę napisać funkcję która odpowiada na pytanie, czy w tablicy każdy wiersz zawiera co najmniej jedną liczbą złożoną wyłącznie z cyfr będących liczbami pierwszymi?


```python
def only_primes(num):
    while num > 0:
        if not is_prime(num % 10):
            return False
        num //= 10
    return True

def only_primes_in_rows(T):
    n = len(T)
    main_flag = True
    for i in range(n):
        secondary_flag = False
        for j in range(n):
            if only_primes(T[i][j]):
                secondary_flag = True
        if secondary_flag == False:
            main_flag = False
    return main_flag

tab = [[222, 444, 444],
       [222, 444, 444],
       [222, 444, 444]]
print(only_primes_in_rows(tab))
```
