### Napisz funkcję, która zwróci sumę elementów w tablicy T

```python
def suma(T, i=0):
    n = len(T)
    if i == n:
        return 0
    return T[i]+suma(T, i+1)


print(suma([1,2,3,4,5]))
```

### Napisz funkcję, która zwróci n-ty element ciągu Fibonacciego

```python
def fib(n):
    if n == 1 or n==2:
        return 1
    return fib(n-1) + fib(n-2)

```

### Napisz funkcję, która stwierdzi czy w tablicy T: bool występuje co najmniej 1 wartość True

```python
def check(T, i=0):
    n = len(T)
    if i == n:
        return False
    return T[i] or check(T, i+1)


print(check([False, False, True]))
```

### Napisz funkcję, która stwierdzi czy w tablicy T: bool występują same wartości True

```python
def check2(T, i=0):
    n = len(T)
    if i == n-1:
        return T[n-i]
    return T[i] and check(T, i+1)

print(check2([True, True, True]))
```
