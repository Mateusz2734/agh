### Napisz funkcję, która zwróci sumę elementów w tablicy T

```python
def suma(T, i=0):
    n = len(T)
    if i == n:
        return 0
    return T[i]+suma(T, i+1)


print(suma([1,2,3,4,5]))
```

### Napisz funkcję, która zwróci n-ty element ciągu Fibonacciego

```python
def fib(n):
    if n == 1 or n==2:
        return 1
    return fib(n-1) + fib(n-2)

```

### Napisz funkcję, która stwierdzi czy w tablicy T: bool występuje co najmniej 1 wartość True

```python
def check(T, i=0):
    n = len(T)
    if i == n:
        return False
    return T[i] or check(T, i+1)


print(check([False, False, True]))
```

### Napisz funkcję, która stwierdzi czy w tablicy T: bool występują same wartości True

```python
def check2(T, i=0):
    n = len(T)
    if i == n-1:
        return T[n-i]
    return T[i] and check(T, i+1)

print(check2([True, True, True]))
```

### Napisz funkcję, która stwierdzi czy w tablicy T: bool występuja 2 wartości True z rzędu

```python
def check3(T, i=0):
    n = len(T)
    if i == n-2:
        return T[n-2] and T[n-1]
    return (T[i] and T[i+1]) or check3(T, i+1)

print(check3([False, True, True]))
```

```python
def A(n):
    return n + 3

def B(n):
    return 2 * n

def C(n):
    px = 1
    x = 10
    while n // px > 0:
        if ((n%x) // px) % 2 == 0:
            n += px
        px = x
        x *= 10
    return n

def zlicz_to(X, Y, N, ost_op):
    if X == Y:
        return 1
    if N == 0:
        return 0
    
    if ost_op == "A":
        return zlicz_to(B(X), Y, N-1, "B") + zlicz_to(C(X), Y, N-1, "C")
    elif ost_op == "B":
        return zlicz_to(A(X), Y, N-1, "A") + zlicz_to(C(X), Y, N-1, "C")
    elif ost_op == "C":
        return zlicz_to(A(X), Y, N-1, "A") + zlicz_to(B(X), Y, N-1, "B")
    else:
        return zlicz_to(A(X), Y, N-1, "A") + zlicz_to(B(X), Y, N-1, "B") + zlicz_to(C(X), Y, N-1, "C")
    
print(zlicz_to(11, 31, 4, ""))
```
