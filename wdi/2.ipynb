### Zadanie 4. Liczba dwu-trzy-piątkowa w rozkładzie na czynniki pierwsze nie posiada innych czynników niż $2,3,5$. Jedynka też jest taką liczbą. Napisz program, który wylicza ile takich liczb znajduje się w przedziale od $1$ do $N$ włącznie.


```python
def isSpecial(n):
    if n == 1:
        return True
    i = 2
    while i <= 5:
        if n % i == 0 and n > 1:
            n = n // i
        else:
            i += 1
    if n == 1:
        return True
    return False
            
def przedzial(N):
    suma = 0
    for i in range(1, N+1):
        if isSpecial(i):
            suma += 1
    return suma

print(przedzial(1000))
```

### Zadanie 6. Napisać program wczytujący liczbę naturalną z klawiatury i rozkładający ją na iloczyn dwóch liczb o najmniejszej różnicy, np. $30 = 5 \cdot 6$, $120 = 10 \cdot 12$.

```python
from math import sqrt

def product(n):
    best1 = 1 
    best2 = n
    for i in range(1, int(sqrt(n))+1):
        if n % i == 0:
            if abs(n//i - i) < abs(best2-best1):
                best1, best2 = i, n//i
    return best1, best2

print(product(120))
```

### Zadanie 9. Napisać program, który oblicza pole figury pod wykresem funkcji $y = \frac{1}{x}$ w przedziale od $1$ do $k$, metodą prostokątów.

```python
def integral(k):
    eps = 1e-6
    x = 1
    area = 0
    while x < k:
        area += eps/x
        x += eps
    return area

print(integral(2))
```

### Zadanie 11. Napisz program wczytujący liczbę naturalną z klawiatury i odpowiadający na pytanie, czy jej cyfry stanowią ciąg rosnący.


```python
def ascending_nums(n):
    last_num = n % 10
    n //= 10
    while n > 0:
        if n % 10 >= last_num:
            return False
        n //= 10
    return True
print(ascending_nums(12345))
```

### Zadanie 12. Napisz program wczytujący liczbę naturalną z klawiatury i odpowiadający na pytanie, czy liczba ta zawiera cyfrę równą liczbie swoich cyfr.

```python
def calc_num_of_digits(n):
    num = 0
    while n > 0:
        n //= 10
        num += 1
    return num

def is_containing(n):
    num_of_digits = calc_num_of_digits(n)
    while n > 0:
        if n % 10 == num_of_digits:
            return True
        n //= 10
    return False

print(is_containing(81111111))
```

```python
def is_last_digit_unique(n):
    last_num = n % 10
    n //= 10
    while n > 0:
        if n % 10 == last_num:
            return False
        n //= 10
    return True

print(is_last_digit_unique(21834))
```

### Zadanie 14. Dane są dwie liczby naturalne z których budujemy trzecią liczbę. W budowanej liczbie muszą wystąpić wszystkie cyfry występujące w liczbach wejściowych. Wzajemna kolejność cyfr każdej z liczb wejściowych musi być zachowana. Na przykład mając liczby $123$ i $75$ możemy zbudować liczby $12375, 17523, 75123, 17253, \ldots$ Proszę napisać funkcję która wyznaczy ile liczb pierwszych można zbudować z dwóch zadanych liczb.

```python
def is_prime(n): 
    if n < 2: 
        return False
    p = int(n**0.5) + 1
    for i in range(2, p):
        if n % i == 0:
            return False
    return True

def leng(n):
    num = 0
    while n > 0:
        n //= 10
        num += 1
    return num

def mix_numbers(a, b, mask):
    num = 0
    mult = 1
    while mask > 0 or a > 0:
        if mask % 2 == 0:
            d = a % 10
            a //= 10
        else:
            d = b % 10
            b //= 10
        mask //= 2
        
        num = d * mult + num
        mult *= 10
    return num

def validate_mask(a, b, mask):
    cnt1 = leng(a)
    cnt2 = leng(b)
    while mask > 0:
        if mask % 2 == 0:
            cnt1 -= 1
        else:
            cnt2 -= 1
        mask //= 2
    return cnt1 >= 0 and cnt2 == 0
    
a = 13
b = 135
counter = 0
for mask in range(2**(leng(a)+leng(b))+1):
    if validate_mask(a,b, mask) and is_prime(mix_numbers(a,b,mask)):
        counter += 1
print(counter)
```

### Zadanie 17. Napisać program wyliczający pierwiastek równania $x^x = 2020$ metodą stycznych.

```python
from math import log

def f(x):
    return x * log(x) - log(2020)

def df(x):
    return log(x) + 1

def solve():
    x = 1
    eps = 1e-6
    while abs(f(x)) > eps:
        x = x - f(x)/df(x)
    return x

print(solve())
```

### Zadanie 20. Dwie liczby naturalne są różno-cyfrowe jeżeli nie posiadają żadnej wspólnej cyfry. Proszę napisać program, który wczytuje dwie liczby naturalne i poszukuje najmniejszej podstawy systemu (w zakresie $2-16$) w którym liczby są różno-cyfrowe. Program powinien wypisać znalezioną podstawę, jeżeli podstawa taka nie istnieje należy wypisać komunikat o jej braku. Na przykład: dla liczb $123$ i $522$ odpowiedzią jest podstawa $11$ bo $123_{(10)} = 102_{(11)}$ i $522_{(10)} = 435_{(11)}$.

```python
def common_digit(x, y, s): #x, y - liczby; s - podstawa systemu
    D = [False] * s
    while x != 0:
        d = x % s
        D[d] = True
        x //= s
    while y != 0:
        if D[y % s]:
            return True
        y //= s
    
def smallest_system(a, b):
    for s in range(2, 17):
        if not common_digit(a, b, s):
            return s
    else:
        return "Brak systemu"

print(smallest_system(123154, 51132))
```
