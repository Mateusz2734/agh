```python
def is_prime(num):
    if num <= 1:
        return False
    if num == 2 or num == 3:
        return True
    if num % 2 == 0 or num % 3 == 0:
        return False
    i = 5
    while i < isqrt(num) + 1:
        if num % i == 0:
            return False
        i += 2
        if num % i == 0:
            return False
        i += 4
    return True

def num_len(n):
    return int(log10(n)) + 1
```

```python
def knights_walk(T, w=0, k=0, n=1):
    T[w][k] = n
    
    if n == len(T) ** 2:
        return True
    
    jumps = [(2, -1), (2, 1), (-2, -1), (-2, 1), (1, -2), (1, 2), (-1, -2), (-1, 2)]
    for jump in jumps:
        next_w = w + jump[0]
        next_k = k + jump[1]
        if contains(T, next_w, next_k) and T[next_w][next_k] == 0:
            if knights_walk(T, next_w, next_k, n+1):
                return True
    T[w][k] = 0
    return False

def contains(T, w, k):
    if 0 <= w < len(T) and 0 <= k < len(T):
        return True
    return False

szachownica = [[0]*6 for _ in range(6)]

print(knights_walk(szachownica))
for row in szachownica:
    print(row)
```

### Zadanie 15. Problem 8 Hetmanów

```python
def print_tab(tab):
    for line in tab:
        print(line)
    print("--------------------")

def queen(lvl, board):
    if lvl == 7:
        print_tab(board)
        return
    for i in range(8):
        if board[lvl][i] == 0:
            bc = copy.deepcopy(board)
            bc[lvl][i] = 2
            k, l = i- 1, lvl+1
            while k > 0 and l < 8:
                bc[l][k] = 1
                k -= 1
                l += 1
            while l < 8:
                bc[l][i] = 1
                l += 1
            k = i+1
            l = lvl + 1
            while k<8 and l<8:
                bc[l][k] = 1
                k += 1
                l += 1
            queen(lvl+1, bc)
            
tab = [[0]*8 for _ in range(8)]
queen(0, tab)
```

Proszę napisać funkcję $wyraz(s_1,s_2)$, która sprawdza czy jest możliwe zbudowanie wyrazu z podzbioru liter zawartych w $s_2$ ważącego tyle co wyraz $s_1$. Dodatkowo funkcja powinna wypisać znaleziony wyraz.

```python
def waga(wyraz):
    samogloski = ["a", "e", "i", "o", "u", "y"]
    samo = 0
    suma = 0
    for litera in wyraz:
        suma += ord(litera)
        if litera in samogloski:
            samo += 1
    return (samo, suma)

def count(s):
    return s.count(elem for elem in ["a", "e", "i", "o", "u", "Y"])

def ascii_sum(s):
    return sum(ord(c) for c in s)

def evaluate(s1, subset):
    if ascii_sum(s1) == ascii_sum(filter(lambda s: s != "", subset)):
        if count(s1) == count(subset):
            return True
    return False
    
    
def rek(s1, s2, i, subset, j):
    if i >= len(s2):
        return evaluate(s1, subset)
    
    subset[j] = s2[i]
    if rek(s1, s2, i+1, subset, j+1):
        return True
    subset[j] = ""
    if rek(s1, s2, i+1, subset, j):
        return True
    
    return False
```

Po tablicy możemy przemieszczać się według następującej zasady: z pola o indeksie $i$ możemy przeskoczyć na pole o indeksie $i+k$ jeżeli $k$ jest czynnikiem pierwszym liczby $T[i]$ mniejszym od $T[i]$. Proszę napisać funkcję, która zwraca informację czy jest możliwe przejście z pola o indeksie $0$ na pole o indeksie $N-1$. Funkcja powinna zwrócić liczbę wykonanych skoków lub wartość $-1$ jeżeli powyższe przejście nie jest możliwe.

```python
def gen_pod(num):
    podzielniki = []
    for i in range(2, isqrt(num)+1):
        if num % i == 0:
            podzielniki.append(i)
            while num % i == 0:
                num //= i
    if num != 1:
        podzielniki.append(num)
    return podzielniki

def zad22(T):
    min_skoki = 0
    def rek(T, maks, i, skoki):
        nonlocal min_skoki
        podzielniki = gen_pod(T[i])
        for podzielnik in podzielniki:
            if i + podzielnik == maks:
                min_skoki = skoki + 1
                return True
            elif i + podzielnik <= maks:
                if rek(T, maks, i+podzielnik, skoki+1):
                    return True
                else:
                    return False
            else:
                return False
    if rek(T, len(T)-1, 0, 0):
        return min_skoki
    return -1
    
T = [2,2,2,2,2]
print(zad22(T))
```

### Zadanie 28. Dany jest zbiór $N$ liczb naturalnych umieszczony w tablicy $T[N]$. Proszę napisać funkcję, która zwraca informację, czy jest możliwy podział zbioru $N$ liczb na trzy podzbiory, tak aby w każdym podzbiorze, łączna liczba jedynek użyta do zapisu elementów tego podzbioru w systemie dwójkowym była jednakowa. Na przykład: $[2, 3, 5, 7, 15] → True$, bo podzbiory $\{2,7\} \{3,5\} \{15\}$ wymagają użycia $4$ jedynek, $[5, 7, 15] → False$, podział nie istnieje.


```python
def ile_jedynek_liczba(num):
    licznik = 0
    while num != 0:
        if num % 2 == 1:
            licznik += 1
        num //= 2
    return licznik

def ile_jed_lis(T):
    licznik = 0
    for liczba in T:
        licznik += ile_jedynek_liczba(liczba)
    return licznik


def zad28(T):
    def rek(T, i, zbior1, zbior2, zbior3):
        if i == len(T):
            if ile_jed_lis(zbior1) == ile_jed_lis(zbior2) == ile_jed_lis(zbior3):
                return True
            return False
        else:
            return rek(T, i+1, zbior1+[T[i]], zbior2, zbior3) or rek(T, i+1, zbior1, zbior2+[T[i]], zbior3) or rek(T, i+1, zbior1, zbior2, zbior3+[T[i]])
    return rek(T, 0, [], [], [])

print(zad28([2,3,5,7,15]))
```

```python
def generuj_podzielniki(num):
    podzielniki = []
    for i in range(2, isqrt(num)+1):
        if num % i == 0:
            podzielniki.append(i)
            while num % i == 0:
                num //= i
    if num != 1:
        podzielniki.append(num)
    return podzielniki

def iloczyn(T):
    il = 1
    for elem in T:
        il *= elem
    return il

def suma_iloczynow(T, suma=0):
    suma = 0
    for podzbior in T:
        suma += iloczyn(podzbior)
    return suma

def zad31(n):
    T = generuj_podzielniki(n)
    podzbiory = []
    def rek(T, i, podzbior):
        if i == len(T):
            if podzbior != []:
                podzbiory.append(podzbior)
                return True
            return False
        else:
            rek(T, i+1, podzbior+[T[i]])
            rek(T, i+1, podzbior)
    rek(T, 0, [])
    return suma_iloczynow(podzbiory)
    

print(zad31(60))
```
