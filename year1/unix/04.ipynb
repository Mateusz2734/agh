### 9. Podaj PID procesu zastopowanego, który zużywa najwięcej pamięci wirtualnej.


```ps --no-headers -eo stat,vsz,pid --sort=vsz | grep "S" | tail -n1 | awk '{print $3}'```

### 10. Podaj nazwę użytkownika, który ma najwięcej procesów z zerową wartością parametru NICE.

```ps --no-headers -eo nice,euser | sort -n | uniq -c | awk '{if ($2 == 0) print $1,$3}' | sort -n | tail -n1```

### 11. Podaj numery PID trzech procesów, które mają najwięcej pamięci wirtualnej.


```ps -eo pid --sort=vsize | tail -n3```

### Jak wyświetlić procesy użytkownika root posortowane wg rozmiaru zużytej pamięci (np. size), używając tylko ps?

```ps -eo uid,pid,%cpu,%mem,cmd,size --sort=+size```

### Ile jest procesów, których ojcem jest proces o PID równym 1?

```ps --no-headers -eo ppid,pid,cmd | awk '{if ($1 == 1) print $0}' | wc -l```

### Znajdź użytkownika w systemie, który ma uruchomioną największą liczbę procesów.

```ps --no-headers -eo uid --sort=uid | uniq -c | sort -n | tail -n1 | awk '{print $2}'```

### Znajdź proces, który zużył najwięcej czasu procesora

```ps --no-headers -eo pid,cputimes --sort=cputimes | tail -n1 | awk '{print $1}'```
