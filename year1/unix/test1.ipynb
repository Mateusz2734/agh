### Ile jest w katalogu /usr/include (oraz wszystkich podkatalogach) plików regularnych, których cała nazwa ma łączną długość 5 znaków, zaczyna się od znaku c, kończy się ciągiem znaków .h i pliki te są większe niż 3 kB?

```find /usr/include -type f -iname "c*" -size +3k | awk -F "/" '{print$NF}' | grep "^...\.h$" | wc -l```

### Ilu użytkowników lokalnie zdefiniowanych w systemie ma numer UID równy lub wyższy od numeru GID grupy podstawowej, do której należą?

```cat /etc/passwd | awk -F ':' '{if ( $3 >= $4 ) print $0}' | wc -l```

### Ile jest w systemie procesów o priorytecie wyższym niż 19 i o numerze PID wyższym niż 800?

```ps --no-headers -eo pid,pri,fname | awk '{if ($2 > 19 && $1 > 800) print $0}' | wc -l```
